## Data Understanding Summary

### Dataset
- 301 rows × 9 columns
- No missing values in any column
- Source: Indian used car market listings
- Prices are in Indian Lakhs (1 Lakh = 100,000 rupees)

### Columns
| Column | Type | Description |
|--------|------|-------------|
| Car_Name | string | Model name — 98 unique values, city is most common (26) |
| Year | int | Manufacturing year — 2003 to 2018 |
| Selling_Price | float | Price the owner is asking (Lakhs) — target variable |
| Present_Price | float | Current ex-showroom price of same model new (Lakhs) |
| Driven_kms | int | Total kilometres driven — wide range (500 to 500,000) |
| Fuel_Type | string | Petrol (79.4%), Diesel (19.9%), CNG (0.7%) |
| Selling_type | string | Dealer (64.8%) vs Individual (35.2%) |
| Transmission | string | Manual (86.7%) vs Automatic (13.3%) |
| Owner | int | Number of previous owners — 0=first, 1=second, 3=fourth |

### Target variable
Selling_Price — ranges from 0.1 to 35 Lakhs, mean 4.66, right-skewed.
Log transformation likely needed during modelling.

### Key observations
- Present_Price is always >= Selling_Price — used cars sell below new price, as expected
- Driven_kms has extreme outliers (max 500,000) — will need investigation in cleaning
- 96.3% of cars are first-owner — Owner column has very limited variance
- CNG fuel type has only 2 entries (0.7%) — may need to merge or drop in cleaning
- Dataset is small (301 rows) — model generalisation will be limited
- 2 fully identical duplicate rows found — will be dropped in cleaning
- 95 shared Car_Name + Year combinations — expected, legitimate separate listings
- Cars span 15 years of manufacture (2003–2018) — age at listing ranges from brand new (0 yrs) to 15 years old
